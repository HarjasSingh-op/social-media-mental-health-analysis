#  Clean the data

```python
health.drop('User_ID', axis=1, inplace=True)


```

```python
health.replace('Male', 0, inplace=True)
health.replace('Female', 1, inplace=True)
health.replace('Other', 2, inplace=True)

health.replace('Facebook', 0, inplace=True)
health.replace('LinkedIn', 1, inplace=True)
health.replace('YouTube', 2, inplace=True)
health.replace('TikTok', 3, inplace=True)
health.replace('X (Twitter)', 4, inplace=True)
health.replace('Instagram', 5, inplace=True)
```

```python
health.head()
```

# Traning and Testing


```python
X, y = health.drop('Happiness_Index(1-10)', axis=1), health["Happiness_Index(1-10)"]

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=87)
```

```python
from sklearn.linear_model import LinearRegression
logmodel = LinearRegression()
logmodel.fit(X_train, y_train)
```

```python
predictions = logmodel.predict( X_test)
logmodel.score(X_test, y_test)
```
